# Balanced parenthesis check

### Gievn a string of opening and closing parentheseis, check whether it is balanced.
### There are 3 types of paranthesis here: Round (), Square brackets [], and curly brackets (braces) {}

_Assumptions_
1. Asssume the string doesn't contain any other character other than the ones menitoned above (no spaces, words, or numbers).
2. Balanced parathesis require every opening parathesis to be closed in the reverse order opened.

```
Examples:
balance_check("([{}])") should return true
balance_check("[[{}])") should return false

Similiarly:
"()()(){{{}}}" is a balanced string
"()()[}((()))" is not
```

#### Instructions
Create a function called "balance_check" that takes a string as its parameter and returns True if it is balanced considering the criteria mentinoed above; False otherwise.

````python
def balance_check(s):
    if len(s) % 2 != 0:
        return False
    opening = set("([{")
    matches = set([
        ("(",")"),
        ("[","]"),
        ("{","}"),
    ])
    stack = []
    for paren in s:
        if paren in opening:
            stack.append(paren)
        else:
            if stack ==[]:
                return False
            last_open = stack.pop()
            if (last_open, paren) not in matches:
                return False
    return stack == []
````

````python
balance_check("()()()[[[]]]")
````

````python
balance_check("()()([][]")
````

````python
class Node:
    def __init__(self, data):
        self.next = None
        self.data = data
````

````python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, element):
        if not self.head:
            self.head = Node(element)
        else:
            current = self.head
            while current.next:   # equivalent to saying "while current.next != None" while current.next exists with data
                current = current.next
            current.next = Node(element)
            
    def insert(self, index, element):
        if not self.head:
            self.head = Node(element)
        else:
            counter = 0
            current = self.head
            prev = None
            while current.next and counter != index:
                prev = current
                current = current.next
                counter += 1
            if index == 0:
                current_head = self.head
                new_node = Node(element)
                new_node.next = current_head
                self.head = new_node
            elif counter != index:
                current.next = Node(element)
            elif counter == index:
                new_node = Node(element)
                prev.next = new_node
                new_node.next = current
    
    def remove(self, value):
        found = False
        if self.head:
            current = self.head
            if current.data == value:
                self.head = self.head.next
                found = True
            else:
                prev = None
                while current.next and current.data != value:
                    prev = current
                    current = current.next
                if current.data == value:
                    prev.next = current.next
                    found = True
        if not found:                                 # if found != True
            raise ValueError("%s not in list" % value)
    
    def __str__(self):
        out = "[ %s" % self.head.data
        current = self.head
        while current:
            current = current.next
            if current:
                out += ", %s" % current.data
        out += "]"
        return out
            
        
    def __len__(self):
        #return length as integer
        if not self.head:
            return 0
        counter = 1
        data = self.head
        while data.next:
            counter = 1+counter
            data = data.next
        return counter
    
    def count(self, value):
        # return the number of times value exists in list
        node = self.head
        counter = 0
        while node is not None:
            if node.data == value:
                counter = counter + 1
                node = node.next
            else:
                node = node.next
        return counter              
        
    def index(self, value):
        # returns first index of "value" if found, ValueError if not found
        active = self.head
        counter = 0
        while active is not None:
            if value == active.data:
                return counter
            counter = counter + 1
            active = active.next
        return ("value does not exist within list")
            
           
````
